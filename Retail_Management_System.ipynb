```python
#class to represent a product in retail management system
class Product:
    #constructor
    def __init__(self, product_id, name, description, price, stock_quantity):#intializing product attributes
        self.product_id = product_id
        self.name = name
        self.description = description
        self.price = price
        self.stock_quantity = stock_quantity
#using getter method for product_id, name,price,stock_quantity
    def get_product_id(self):
        return self.product_id

    def get_name(self):
        return self.name

    def get_description(self):
        return self.description

    def get_price(self):
        return self.price

    def get_stock_quantity(self):
        return self.stock_quantity
#using setter method for name,description,price,stock_quantity
    def set_name(self, name):
        self.name = name

    def set_description(self, description):
        self.description = description

    def set_price(self, price):
        self.price = price

    def set_stock_quantity(self, stock_quantity):
        self.stock_quantity = stock_quantity
#method
    def add_product(self, products):#addind product to product list 
        products.append(self)
        print("Product added successfully!")
#updating product attributes
    def update_product(self, name, description, price, stock_quantity):
        self.name = name
        self.description = description
        self.price = price
        self.stock_quantity = stock_quantity
        print("Product updated successfully!")
#checking stock_quantity
    def check_stock(self):
        return self.stock_quantity
#generating product detail report
    def generate_product_report(self):
        return f"Product ID: {self.product_id}\nName: {self.name}\nDescription: {self.description}\nPrice: {self.price}\nStock Quantity: {self.stock_quantity}"

#class representing customer in retail system 
class Customer:
    def __init__(self, customer_id, name, email):
        self.customer_id = customer_id
        self.name = name
        self.email = email
        self.purchase_history = []
#using getter method for customer_id,name,email,purchase_history
    def get_customer_id(self):
        return self.customer_id

    def get_name(self):
        return self.name

    def get_email(self):
        return self.email

    def get_purchase_history(self):
        return self.purchase_history
#using setter method for name,email,register_customer
    def set_name(self, name):
        self.name = name

    def set_email(self, email):
        self.email = email
#method for registering customers
    def register_customer(self, customers):
        customers.append(self)
        print("Customer registered successfully!")
#updating customer attributes
    def update_customer(self, name, email):
        self.name = name
        self.email = email
        print("Customer updated successfully!")
#adding purchase to customer purchasing history
    def add_purchase(self, sale):
        self.purchase_history.append(sale)
#generating report for customer details and purchase history
    def generate_customer_report(self):
        report = f"Customer ID: {self.customer_id}\nName: {self.name}\nEmail: {self.email}\nPurchase History:\n"
        for sale in self.purchase_history:
            report += f"\tSale ID: {sale.get_sale_id()}, Product: {sale.get_product().get_name()}, Quantity: {sale.get_quantity()}, Total Price: {sale.get_total_price()}\n"
        return report

#class representing sales in management retail system
class Sale:
    def __init__(self, sale_id, customer, product, quantity):
        self.sale_id = sale_id
        self.customer = customer
        self.product = product
        self.quantity = quantity
        self.total_price = product.get_price() * quantity
#using getter method for sale,customer,product,quantity,total_price
    def get_sale_id(self):
        return self.sale_id

    def get_customer(self):
        return self.customer

    def get_product(self):
        return self.product

    def get_quantity(self):
        return self.quantity

    def get_total_price(self):
        return self.total_price
#using setter method for customer ,product,quantity,total_price
    def set_customer(self, customer):
        self.customer = customer

    def set_product(self, product):
        self.product = product

    def set_quantity(self, quantity):
        self.quantity = quantity

    def set_total_price(self, total_price):
        self.total_price = total_price
#method for processing salesw
    def process_sale(self):
        if self.product.get_stock_quantity() >= self.quantity:
            self.product.set_stock_quantity(self.product.get_stock_quantity() - self.quantity)
            self.customer.add_purchase(self)
            print("Sale processed successfully!")
        else:
            print("Error: Not enough stock!")
#generating reports for sales detail
    def generate_sales_report(self):
        return f"Sale ID: {self.sale_id}\nCustomer: {self.customer.get_name()}\nProduct: {self.product.get_name()}\nQuantity: {self.quantity}\nTotal Price: {self.total_price}"

#function for generating all product reports
def generate_product_report(products):
    report = ""
    for product in products:
        report += product.generate_product_report() + "\n\n"
    print("Product Report:\n" + report)

#generating report of all customers
def generate_customer_report(customers):
    report = ""
    for customer in customers:
        report += customer.generate_customer_report() + "\n\n"
    print("Customer Report:\n" + report)
#generating all sales report

def generate_sales_report(sales):
    report = ""
    for sale in sales:
        report += sale.generate_sales_report() + "\n\n"
    print("Sales Report:\n" + report)

#main function for processing of retail management system
if __name__ == "__main__":
    #list to store product ,customers,sales
    products = []
    customers = []
    sales = []

    while True:
        print("\nRetail Management System")
        print("1. Add Product")
        print("2. Register Customer")
        print("3. Process Sale")
        print("4. Generate Reports")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            #addition of new product
            product_id = input("Enter Product ID: ")
            name = input("Enter Product Name: ")
            description = input("Enter Product Description: ")
            price = float(input("Enter Product Price: "))
            stock_quantity = int(input("Enter Stock Quantity: "))
            product = Product(product_id, name, description, price, stock_quantity)
            product.add_product(products)

        elif choice == '2':
            #registeration of new customer
            customer_id = input("Enter Customer ID: ")
            name = input("Enter Customer Name: ")
            email = input("Enter Customer Email: ")
            customer = Customer(customer_id, name, email)
            customer.register_customer(customers)

        elif choice == '3':
            #processing of new sales
            sale_id = input("Enter Sale ID: ")
            customer_id = input("Enter Customer ID: ")
            product_id = input("Enter Product ID: ")
            quantity = int(input("Enter Quantity: "))
#finding customer and product by id 
            customer = next((c for c in customers if c.get_customer_id() == customer_id), None)
            product = next((p for p in products if p.get_product_id() == product_id), None)


            if customer and product:
                sale = Sale(sale_id, customer, product, quantity)
                sale.process_sale()
                sales.append(sale)
            else:
                print("Error: Customer or Product not found!")

        elif choice == '4':
            #generating reports
            print("\nGenerate Reports")
            print("1. Generate Product Report")
            print("2. Generate Customer Report")
            print("3. Generate Sales Report")
            report_choice = input("Enter your choice: ")

            if report_choice == '1':
                generate_product_report(products)
            elif report_choice == '2':
                generate_customer_report(customers)
            elif report_choice == '3':
                generate_sales_report(sales)
            else:
                print("Invalid choice!")

        elif choice == '5':
            break# this exists the system

        else:
            print("Invalid choice! Please try again.")
```
